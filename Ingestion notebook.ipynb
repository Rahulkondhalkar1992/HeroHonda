```python
dbutils.widget.text("db", "db")
dbutils.widgets.text("path", "path")
dbutils.widgets.text("file", "file")
dbutils.widgets.text("table", "table"
```

```python
#write for loop for hero honda data
for i in range(1,10):
  df = spark.read.format("csv").option("header", "true").option("inferSchema", "true").load(f"/FileStore/tables/{dbutils.widgets.get("db")}/{dbutils.widgets.get("path")}/{dbutils.widgets. get("file")}{i}.csv")
  df.write.mode("append").saveAsTable(f"{dbutils.widgets.get("table")}")        
```
